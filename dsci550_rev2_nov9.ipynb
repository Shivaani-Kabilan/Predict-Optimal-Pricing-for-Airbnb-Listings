```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

#load dataset
df= pd.read_csv("listings.csv")
df
```

```python
df1=df.loc[df['neighbourhood_group_cleansed'] == "City of Los Angeles"]
df1
```

```python
#drop NA values for some of the numerical data columns
#df2=df1.dropna(subset=['review_scores_value', 'reviews_per_month', 'latitude', 'longitude', 'price'])
df2=df1.dropna(subset=['price'])
#convert price from string to int
df2.loc[:,"price"] = df2.loc[:,"price"].str.replace('[\$\,]', '').astype(float)
df2.head()
```

```python
#selecting meaningful features to predictt price
df3=df2[['name', 'neighborhood_overview', 'neighbourhood', 'latitude', 'longitude', 'property_type', 'room_type', 'bedrooms', 'beds', 'bathrooms_text', 'amenities', 'availability_365',  'review_scores_value', 'reviews_per_month', 'price' ]]
df3.head()
```

```python
impute_median=df['bedrooms'].median()
df3['bedrooms']=df3['bedrooms'].fillna(impute_median)
df3['beds']=df3['beds'].fillna(impute_median)
df3['review_scores_value']=df3['review_scores_value'].fillna(impute_median)
df3['reviews_per_month']=df3['reviews_per_month'].fillna(impute_median)
df3['latitude']=df3['latitude'].fillna(impute_median)
df3['longitude']=df3['longitude'].fillna(impute_median)


df3.head()
```

```python
import re
df3["bathrooms_text"]=df3["bathrooms_text"].apply(str)

def find_number(text):
    f=0
    match = re.search(r'[0-9]+', text)
    if match:
        f=1
        return match.group()
    else:
        return f

#def find_number(text):
    #num = re.findall(r'[0-9]+',text)
    #s=" ".join(num)[:1]
    #return float(s)
df3['bathrooms_text']=df3['bathrooms_text'].apply(lambda x: find_number(x))
df3.head()
```

```python
df3.sort_values(by=['price']).tail(5)
```

```python
df3 = df3[df3["room_type"].str.contains("Hotel room") == False]
df3.sort_values(by=['price']).tail(5)
df2=df1.dropna(subset=['price'])
```

```python
import pandas as pd
from sklearn import linear_model

X = df3[['latitude', 'longitude', 'availability_365', 'bedrooms', 'bathrooms_text', 'reviews_per_month', 'review_scores_value', 'beds']]
Y = df3['price']

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split( X, Y, test_size=0.2, random_state=0)
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train,Y_train)

y_pred= regressor.predict(X_test)
print(y_pred)

regr = linear_model.LinearRegression()
regr.fit(X, Y)

print(regr.coef_)

import sklearn.metrics as metrics

mae = metrics.mean_absolute_error(Y_test, y_pred)
mse = metrics.mean_squared_error(Y_test, y_pred)
rmse = np.sqrt(mse) # or mse**(0.5)  
r2 = metrics.r2_score(Y_test,y_pred)

print("Results of sklearn.metrics:")
print("MAE:",mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R-Squared:", r2)
```
